```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Resources/amazon.csv")
df
```

```python
df = df.dropna()
df
```

```python
df.loc[:, 'discounted_price'] = df['discounted_price'].replace(',', '', regex=True)
df.loc[:, 'discounted_price'] = df['discounted_price'].replace('₹', '', regex=True)
```

```python
df
```

```python
df.loc[:, 'actual_price'] = df['actual_price'].replace(',', '', regex=True)
df.loc[:, 'actual_price'] = df['actual_price'].replace('₹', '', regex=True)
```

```python
df["actual_price"] = df["actual_price"].astype(float)
```

```python
df.dtypes
```

```python
df["discounted_price"] = df["discounted_price"].astype(float)
```

```python
df.dtypes
```

```python
df
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(float)
df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(float)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

features = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
target_variable = 'discounted_price'

x = df[features].drop(target_variable, axis=1)
y = df[target_variable]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
fit_model = model.fit(x_train, y_train)

predictions = model.predict(x_test)

mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
r_squared = r2_score(y_test, predictions)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r_squared}")
```

```python
import matplotlib.pyplot as plt
# Convert y_test and predictions to NumPy arrays
y_test_array = y_test.values.astype(float)
predictions_array = predictions.astype(float)

# Plotting the actual vs. predicted values as distinct points
plt.figure(figsize=(8, 6))

# Scatter plot of actual vs. predicted values
plt.scatter(y_test_array, predictions_array, alpha=0.5, label='Actual vs. Predicted', c='blue')

# Plot the diagonal line (best-fit line)
y_min = min(np.min(y_test_array), np.min(predictions_array))
y_max = max(np.max(y_test_array), np.max(predictions_array))
plt.plot([y_min, y_max], [y_min, y_max], color='red', label='Best-fit line')

plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.legend()
plt.grid(True)
plt.show()
```
